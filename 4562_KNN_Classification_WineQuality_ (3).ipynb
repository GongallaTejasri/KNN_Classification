# **problem statement**

The Wine Quality Dataset is a collection of data on the physicochemical 
properties and sensory quality of red and white wine variants from portugal .
The dataset contains 12 features or variables, including 11 input variables that describe various physicochemical properties of the wine and one output variable that represents the sensory quality of the wine rated by human experts on a scale from 0 to 10 .The dataset includes a total of 2056 observations .


The 11 input variables in the dataset are:

1. Fixed acidity: The amount of fixed acids in the wine.
2. Volatile acidity : The amount of volatile acids in the wine.
3. Citric acid: The amount of citric acid in the wine.
4. Residual sugar: The amount of residual sugar in the wine.
5. Chloraides: The amount of salts in the Wine.
6. Free sulfur dioxide: The amount of free sulfur dioxide in the wine.
7. Total sulfur dioxide: The amount of total sulfur dioxide in the wine.
8. Density: The density of the Wine.
9.  pH: The pH level of the Wine.
10. Alcohol: The alcohol content of the Wine.

The out variable is:

1. Quality : The sensory quality of the Wine rated by human experts on a scale from 0 to 10.

```python
# import the random module 
import random

#define a function train_test_split that takes a DataFrame 'X' and a test size as input
def train_test_split(X,test_size=0.2):
  
  #get the size of the input DataFrame 'X'
  Wine_Quality_size = len(X)
  
  #calculate the size of the test data based on the test size parameter
  Test_Data_size = int(Wine_Quality_size*test_size)

  #calculate the size of the train data by substracting the test data size from the input data size
  Train_Data_size = Wine_Quality_size - Test_Data_size
  
  #function Random_shuffle to shuffle the input data randomly
  def Random_shuffle(X):
    for i in range(len(X)-1,0,-1):
      j = random.randint(0,i)
      X[i],X[j] = X[j],X[i]
  
  #split the shuffled data in to train and test data based on the calculated train and test data sizes
  X_train=X[:Train_Data_size]
  X_test = X[Train_Data_size:]

  # return the train and test data as separate DataFrames
  return X_train,X_test
```

```python
class StandardScaler:
    def fit(X):
        mean_ = np.mean(X, axis=0)# Calculate the mean of each column of X
        std_ = np.std(X, axis=0)
        return mean_, std_ # Return the mean and standard deviation as a tuple

    def transform(X, mean_, std_):
        X_std = (X - mean_) / std_# Standardize the input data X using the mean and standard deviation
        return X_std# Return the standardized data

    def fit_transform(X):
        mean_, std_ = StandardScaler.fit(X)# Calculate the mean and standard deviation of X
        X_std = StandardScaler.transform(X, mean_, std_)# Standardize the input data X using the mean and standard deviation
        return X_std# Return the standardized data
```

## **KNN Classification**

```python
import numpy as np


class KNNClassifier:
    def __init__(self, k):
        self.k = k#constructor method that initilazies the k value

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):#Method to make predictions on the test data
        Output = np.zeros(X_test.shape[0])
        for i, x_test in enumerate(X_test):
          distances =  dist = np.power(np.sum(np.power(np.abs(X_train[i] - X_test), 1)), 1/1) # Calculate distances between the test point and all training points
            # Get the k nearest neighbors
        nn_indices = np.argsort(distances)[:self.k]
        nn_labels = self.y_train[nn_indices]
        # Get the most common class label among the neighbors
        Output[i] = np.bincount(nn_labels).argmax()
        
        return Output
```

```python
knn =KNNClassifier(k=5) #create an instance of the KNNClassifier with k=5
knn.fit(X_train,y_train)#fit the training  data(X_train and y_train) to the KNN model
```

# **Accuracy**

```python
def model_accuracy(y_test,Output):
    correct = 0
    # Counter to track of number of correct predictions
    for i in range(len(y_test)):
      # Loop over each element in y_test and Output
      if y_test[i]==Output[i]:
        # check if prediction is correct
        correct +=1
    # calculate and return the accuracy score
    return correct/len(y_test)
```

```python
accuracy = model_accuracy(y_test,Output)
# calculate the accuracy score by comparing y_test and Output

print("Accuracy:",accuracy*100)
#print the accuracy score as a percentage
```
